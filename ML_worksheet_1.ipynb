#Find average SOLD PRICE for each age category, group all records by AGE and then apply mean() on SOLD PRICE column

# Perform joining of data frame- compare the average SOLD PRICE for different AGE categories with the different age and PLAYING ROLE categories, we need to merge the DataFrames soldprice_by_age and soldprice_by_ age_role. The common column is AGE and this needs outer join. Age is considered as categorical varaibale(as 1,2,3)

```python
# prompt: Perform joining of data frame- compare the average SOLD PRICE for different AGE categories with the different age and PLAYING ROLE categories, we need to merge the DataFrames soldprice_by_age and soldprice_by_ age_role. The common column is AGE and this needs outer join. Age is considered as categorical varaibale(as 1,2,3)

import pandas as pd

# Create the DataFrames
soldprice_by_age = df.groupby('AGE')['SOLD PRICE'].mean()
soldprice_by_age_role = df.groupby(['AGE', 'PLAYING ROLE'])['SOLD PRICE'].mean()

# Merge the DataFrames
merged_df = pd.merge(soldprice_by_age, soldprice_by_age_role, on='AGE', how='outer')

# Print the merged DataFrame
print(merged_df)
```

# To find whether players carry a premium if they belong to a specific AGE and PLAYING ROLE category. The premium (we will call it change) is calculated in percentage terms and calculated as follows


change=(Average Sold Price for all player in an AGE and PLAYING Role category - Average SOLD PRICE for all player in an AGE category)/(Average SOLD PRICE for all player in an AGE category)


```python
# prompt: To find whether players carry a premium if they belong to a specific AGE and PLAYING ROLE category. The premium (we will call it change) is calculated in percentage terms and calculated as follows
# change=(Average Sold Price for all player in an AGE and PLAYING Role category - Average SOLD PRICE for all player in an AGE category)/(Average SOLD PRICE for all player in an AGE category)

merged_df['change'] = (merged_df['SOLD PRICE_y'] - merged_df['SOLD PRICE_x']) / merged_df['SOLD PRICE_x'] * 100

print(merged_df)
```

# Exploration of data using visualization-Drawing Plots -Matplotlib -Bar Chart, scatter plot Histogram, Correlation and Heatmap.

```python
# prompt: Exploration of data using visualization-Drawing Plots -Matplotlib -Bar Chart, scatter plot Histogram, Correlation and Heatmap.

import matplotlib.pyplot as plt
import seaborn as sns

# Bar chart
df['AGE'].value_counts().plot(kind='bar')
plt.title('Bar chart of AGE')
plt.xlabel('AGE')
plt.ylabel('Count')
plt.show()

# Scatter plot
plt.scatter(df['RUNS-S'], df['SOLD PRICE'])
plt.title('Scatter plot of RUNS-S vs SOLD PRICE')
plt.xlabel('RUNS-S')
plt.ylabel('SOLD PRICE')
plt.show()

# Histogram
plt.hist(df['AGE'], bins=5)
plt.title('Histogram of AGE')
plt.xlabel('AGE')
plt.ylabel('Count')
plt.show()

# Correlation heatmap
corr = df[['RUNS-S', 'SOLD PRICE']].corr()
sns.heatmap(corr, annot=True)
plt.title('Correlation heatmap')
plt.show()
```
